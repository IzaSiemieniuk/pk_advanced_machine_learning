```python
import pandas as pd
```

```python
df = pd.read_csv('data/pet_adoption_data.csv')
```

```python
df.head()
```

```python
columns = ['PetType', 'AgeMonths', 'Size','HealthCondition', 'AdoptionFee']
percentage = [0.02, 0.02, 0.03, 0.015, 0.01]
```

```python
for i, (col, perc) in enumerate(zip(columns, percentage)):
    df.loc[df.sample(frac=perc, random_state=102 + i).index, col] = pd.NA
```

```python
df.notna().mean().mean()
```

```python
df.notna().mean()
```

```python
df.isna().any(axis=1).mean() * 100
```

```python
df.to_csv('data/pet_adoption_data_with_missing.csv', index=False)
```
